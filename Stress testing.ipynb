```python
import re
import requests
from requests.auth import HTTPBasicAuth
import json
import base64

admin_username = "admin"
admin_password = "ClXciOrtWMYtzYob"
auth = HTTPBasicAuth(admin_username, admin_password)

base_url = "https://nsspc.ntu.edu.tw/domjudge"
headers = {
    'Content-Type': 'application/json',
}    

def login(user, password = None):
    session = requests.Session()

    # Get CSRF token
    login_page = session.get(f"{base_url}/login")
    csrf_token = re.search(r'<input type="hidden" name="_csrf_token" value="([^"]*)">', login_page.text).group(1)

    # Login
    payload = {
        "_username": user,
        "_password": password if password else user,
        "_csrf_token": csrf_token
    }

    login_response = session.post(f"{base_url}/login", data = payload)

    if login_response.status_code == 200:
        print(f"{user} logged in successfully.")
    else:
        raise Exception(f"Failed to login, user: {user}.")

def add_user(data):
    # Make the request with basic auth
    response = requests.post(f"{base_url}/api/users", auth = auth, data = json.dumps(data), headers = headers)
    
    # Check the response
    if response.status_code in [200, 201]:
        print('Success:', response.json())
    else:
        raise Exception('Failed:', response.status_code, response.text)

def add_team(data):
    # Make the request with basic auth
    response = requests.post(f"{base_url}/api/contests/1/teams", auth = auth, data = json.dumps(data), headers = headers)
    
    # Check the response
    if response.status_code in [200, 201]:
        print('Success:', response.json())
        return response.json()
    else:
        raise Exception('Failed:', response.status_code, response.text)

def submit(user, password, data):
    user_auth = HTTPBasicAuth(user, password)

    # Make the request with basic auth
    response = requests.post(f"{base_url}/api/contests/1/submissions", auth = user_auth, data = json.dumps(data), headers = headers)
    
    # Check the response
    if response.status_code in [200, 201]:
        print('Success:', response.content)
        return response.json()
    else:
        raise Exception('Failed:', response.status_code, response.text)

login(admin_username, admin_password)


# files = []

# with open("test.zip", 'rb') as pyfile:
#     file_content = pyfile.read()
# file = base64.b64encode(file_content).decode('utf-8')

# files.append({"data": file, "mime": "application/zip"})
# submit("test_team1_user1", "1111111111", {
#     "language_id": "python3",
#     "problem_id": "1",
#     "team_id": "22",
#     "files": files,
# })
# trial = 2
# batch = 250
# member = 3

# for i in range(1 + batch * (trial - 1), 1 + batch * trial):
#     for j in range(1, member + 1):
#         login(f"test_team{i}_user{j}", str(j) * 10)
```
